TUTOTRIAL 1 (SIMPLE SINGLE LAYER PERCEPTRON)

```python
import numpy as np
```

```python
# GIVEN DATA
x = np.array([1.0, 2.0]) # INPUTS
y = 1                    # TARGETS
w = np.array([0.2, -0.5])      # INITIAL WIEGHTS
b = 0.1                  # Intial bias
eta = 0.1                # learning rate

# To connect this notebook with GitHub:
# 1. To save a copy to GitHub: File -> Save a copy in GitHub
# 2. To open a notebook from GitHub: File -> Open notebook -> GitHub (tab)
```

```python
# run for 3 epochs
for epoch in range(1, 4):
  print(f"\nEpoch {epoch}:")
  net_input = np.dot(w, x) + b    # we taking dot product
  y_pred = 1 if net_input > 0 else 0 # this is step activation function
  error= y - y_pred

  print(f"Net Input: {net_input}")
  print(f"Predicted Output: {y_pred}")
  print(f"Error: {error}")


  # updating rule
  w += eta * error * x
  b += eta * error

  print(f" Updated Weights: {w}")
  print(f" Updated Bias: {b}" )
```
